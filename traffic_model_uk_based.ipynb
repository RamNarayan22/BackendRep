## Handle missing values

Address the missing values in columns like 'Carriageway_Hazards' and 'Time'.


**Reasoning**:
Handle missing values in 'Carriageway_Hazards' by filling with the mode and drop rows with missing 'Time' values. Then verify that missing values in these columns are handled.



```python
# Handle missing values in 'Carriageway_Hazards' by filling with the mode
mode_carriageway_hazards = data['Carriageway_Hazards'].mode()[0]
data['Carriageway_Hazards'] = data['Carriageway_Hazards'].fillna(mode_carriageway_hazards)

# Drop rows with missing 'Time' values
data = data.dropna(subset=['Time'])

# Verify that missing values in 'Carriageway_Hazards' and 'Time' have been handled
print(data[['Carriageway_Hazards', 'Time']].isnull().sum())
```

## Encode categorical columns

### Subtask:
Apply label encoding or other appropriate encoding techniques to the remaining categorical columns.


**Reasoning**:

```python
# Identify object type columns
object_cols = data.select_dtypes(include='object').columns

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to each object type column
for col in object_cols:
    data[col] = le.fit_transform(data[col])

# Display the first few rows to see the encoded columns
display(data.head())
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

# Define the neural network model
model = Sequential()

# Add the input layer
model.add(Dense(units=64, activation='relu', input_shape=(y_train.shape[1],)))

# Add a hidden layer
model.add(Dense(units=32, activation='relu'))

# Add the output layer (assuming 3 severity levels based on previous output)
num_classes = len(x_train.unique())
model.add(Dense(units=num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(y_train, x_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model on the test data
loss, accuracy = model.evaluate(y_test, x_test)
print(f"Test Accuracy: {accuracy}")
```
